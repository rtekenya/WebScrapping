```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
url = 'https://superbalist.com/browse/men/shoes/sneakers?designer_s%5B0%5D=adidas-originals'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

#grabs each product
products = soup.findAll("div", {"class": "bucket bucket-product-with-details bucket-product-with-attributes"})
#print(products)

#open file name
filename = "adidas.csv "
f = open(filename, "a")
headers = "brand, price \n"
f.write(headers)

count = 0
for brand_name in products:
    brand1 = brand_name.findAll("h4", {"class": "product-title"})
    brand = brand1[0].text.strip()
    #print(count, brand)
    
    prices = brand_name.findAll("span", {"class": "price"})
    price = prices[0].text.strip()
    count += 1
    print(count, brand + ":" + price)
    f.write(brand + ":" + price  + "\n")
f.close()
```
